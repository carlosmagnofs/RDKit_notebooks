```python
from rdkit import Chem
from rdkit.Chem import Draw
from rdkit.Chem import AllChem
```

```python
mol1 = Chem.MolFromSmiles('CC(=O)O')
img1 = Draw.MolToImage(mol1)
img1
```

```python
subg1 = Chem.MolFromSmarts('C(=O)[OH]')            #Utilização dos SMARTS nos subgrupos
rmol1 = AllChem.DeleteSubstructs(mol1,subg1)       #Deleta o subgrupo indicado
imgr1 = Draw.MolToImage(rmol1)
imgr1

```

```python
mol2 = Chem.MolFromSmiles('CCC(C1=CC=CC=C1)CCN')
img2 = Draw.MolToImage(mol2)
img2
```

```python
subg2 = Chem.MolFromSmarts('cccccc')
rmol2 = AllChem.DeleteSubstructs(mol2,subg2)
imgr2 = Draw.MolToImage(rmol2)
imgr2
```

```python
mol3 = Chem.MolFromSmiles('CCNC(C1=CC=CC=C1)CCN')
img3 = Draw.MolToImage(mol3)
img3
```

```python
subg3 = Chem.MolFromSmarts('N')                #Em SMARTS as strings fora dos parênteses são interpretadas de forma generalizada por padrão
rmol3 = AllChem.DeleteSubstructs(mol3,subg3)
imgr3 = Draw.MolToImage(rmol3)
imgr3
```

```python
subg3 = Chem.MolFromSmarts('[NH2]')             #AS strings dentro dos colchetes são interpretadas de forma específica
rmol3 = AllChem.DeleteSubstructs(mol3,subg3)
imgr3 = Draw.MolToImage(rmol3)
imgr3
```

```python
subg4 = Chem.MolFromSmarts('[!C]')                #O ! indica "não" e o C indica carbonos alifáticos
rmol4 = AllChem.DeleteSubstructs(mol3,subg4)
imgr4 = Draw.MolToImage(rmol4)
imgr4
```

```python
mol5 = Chem.MolFromSmiles('c1cc(CCCO)ccc1')
img5 = Draw.MolToImage(mol5)
img5
```

```python
subg5 = Chem.MolFromSmarts('CO')
tr5 = Chem.MolFromSmiles('C(=O)O')
rmol5 = Chem.ReplaceSubstructs(mol5, subg5, tr5)            #Substitui o subgrupo indicado pelo novo subgrupo indicado
imgr5 = Draw.MolToImage(rmol5[0])
imgr5
```

```python
mol6 = Chem.MolFromSmiles('BrCCc1cncnc1C(=O)O')
img6 = Draw.MolToImage(mol6)
img6
```

```python
core = Chem.MolFromSmiles('c1cncnc1')
rmol6 = Chem.ReplaceSidechains(mol6 ,core)        #Identifica uma cadeia como principal e deleta as cadeias restantes
imgr6 = Draw.MolToImage(rmol6)
imgr6
```

```python
rmol7 = Chem.ReplaceCore(mol6,core)              #Deleta a cadeia principal deixando apenas as cadeias adjacentes
imgr7 = Draw.MolToImage(rmol7)
imgr7
```
